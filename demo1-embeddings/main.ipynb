# Demo 1 — Sentence Embeddings
Load the model once, then run each step below.

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("all-MiniLM-L6-v2")
```

## Step 1 — Generate embeddings for the sentences

```python
sentences = [
    "How old are you?",
    "What is your age?",
    "The weather is nice today.",
]

sentence_embeddings = model.encode(sentences, convert_to_tensor=True)

for sentence, vector in zip(sentences, sentence_embeddings):
    print(f"{sentence}")
    print(len(vector), vector[0:10])
    print()
```

## Step 2 — Generate embedding for the query

```python
query = "Do you thing it's gonna rain?"
query_embedding = model.encode(query, convert_to_tensor=True)

print(len(query_embedding), query_embedding[0:10])
```

## Step 3 — Compute similarity scores

```python
scores = util.cos_sim(query_embedding, sentence_embeddings)[0]

for sentence, score in zip(sentences, scores):
    print(f"{score.item():.3f}\t{sentence}")
```

## Step 4 — Sort by score

```python
results = sorted(zip(sentences, scores), key=lambda x: x[1].item(), reverse=True)

for sentence, score in results:
    print(f"{score.item():.3f}\t{sentence}")
```
